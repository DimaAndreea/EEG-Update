```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Incarcarea fisierelor de date

Un fisier contine datele EEG brute cu timestamp-uri, in timp ce celalalt contine informatii despre stimuli, de asemenea cu timestamp-uri.


```python
triggers_df = pd.read_csv('results/emotiv-18-06-2025_20-46-50_triggers.csv')

eeg_data_df = pd.read_csv('results/emotiv-18-06-2025_20-46-50_eeg_data.csv')
```

### Ajustarea timestamp-urilor trigger

Aliniem timestamp-urile trigger-ului cu cele din datele EEG, astfel incat sa fie sincronizate.


```python
first_eeg_timestamp = eeg_df['Timestamp'][0]

trigger_df['Adjusted_Timestamp'] = trigger_df['Timestamp'] - (trigger_df['Timestamp'].iloc[0] - first_eeg_timestamp)

print(trigger_df[['Stimulus', 'Trigger Type', 'Adjusted_Timestamp']].head())
```

### Extragem datele EEG pentru triggere

Pentru fiecare trigger, am extras datele EEG intr-un interval de ±600ms in jurul timestamp-ului trigger-ului. Tipul stimulului este adaugat la datele EEG pentru fiecare trigger. Datele extrase sunt stocate intr-o lista si apoi combinate intr-un singur DataFrame.


```python
interval_ms = 600
interval_sec = interval_ms / 1000

all_triggered_eeg_data = []

for index, row in trigger_df.iterrows():
    stimulus_time = row['Adjusted_Timestamp']
    stimulus = row['Stimulus']

    triggered_eeg = eeg_df[
        (eeg_df['Timestamp'] >= stimulus_time - interval_sec) & 
        (eeg_df['Timestamp'] <= stimulus_time + interval_sec)
    ].copy()

    triggered_eeg.loc[:, 'Stimulus'] = stimulus  

    all_triggered_eeg_data.append(triggered_eeg)

    print(f"Stimulus Timestamp: {stimulus_time}")
    print(f"Interval căutare: [{stimulus_time - interval_sec}, {stimulus_time + interval_sec}]")
    print(f"Număr de date EEG extrase: {len(triggered_eeg)}")

all_triggered_data = pd.concat(all_triggered_eeg_data, ignore_index=True)

print(all_triggered_data.head())
```

### Calcularea mediei semnalului EEG pentru stimuli

Am selectat doar coloanele numerice din cadrul datelor EEG si am calculat media pt fiecare canal pt stimulii "happy" si "sad". Rezultatele sunt afisate pt fiecare stimul in parte.


```python
numeric_columns = all_triggered_data.select_dtypes(include='number')

happy_eeg_data = all_triggered_data[all_triggered_data['Stimulus'] == 'happy']
sad_eeg_data = all_triggered_data[all_triggered_data['Stimulus'] == 'sad']

happy_avg = happy_eeg_data[numeric_columns.columns].mean()  
sad_avg = sad_eeg_data[numeric_columns.columns].mean() 

print("Medii pt stimul vesel:")
print(happy_avg)

print("\nMedii pt stimul trist:")
print(sad_avg)
```

### Vizualizarea semnalelor EEG pentru stimuli veseli si tristi

Extrag datele EEG pentru stimulii mei si afisez semnalele pentru fiecare canal EEG sub forma unor subgrafice. 


```python
happy_data = all_triggered_data[all_triggered_data['Stimulus'] == 'happy']
sad_data = all_triggered_data[all_triggered_data['Stimulus'] == 'sad']

channels = happy_data.columns[1:-1]  
fig, axes = plt.subplots(len(channels), 1, figsize=(10, 2 * len(channels)))

if len(channels) == 1:
    axes = [axes]

for i, channel in enumerate(channels):
    ax = axes[i]
    
    ax.plot(happy_data['Timestamp'], happy_data[channel], label='Happy Stimulus', color='b')
    
    ax.plot(sad_data['Timestamp'], sad_data[channel], label='Sad Stimulus', color='r')
    
    ax.set_title(f'ERP for {channel}')
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('EEG Signal')
    ax.legend()

plt.tight_layout()
plt.show()
```
