# EEG Data Analysis with MNE-Python

This notebook details the process of analyzing EEG data collected from the headset, along with associated stimulus triggers. The goal is to extract Event-Related Potentials (ERPs) by segmenting and averaging EEG activity around specific stimuli.

```python
import pandas as pd
import numpy as np
import mne
import matplotlib.pyplot as plt
import os
```

## Step 1: Define Experiment Parameters and Load Data

This section defines the sampling frequency of the EEG device and specifies the file paths for the collected EEG and trigger data. These CSV files are **then** loaded into Pandas DataFrames.

```python
sfreq = 128  
tmin, tmax = -0.6, 0.6 # -600ms before trigger, +600ms after trigger

eeg_file_name = 'results/emotiv-25-06-2025_13-58-36_eeg_data.csv'
triggers_file_name = 'results/emotiv-25-06-2025_13-58-36_triggers.csv'

if not os.path.exists(eeg_file_name):
    print(f"Error: EEG file '{eeg_file_name}' not found.")
elif not os.path.exists(triggers_file_name):
    print(f"Error: Triggers file '{triggers_file_name}' not found.")
else:
    print(f"Files '{eeg_file_name}' and '{triggers_file_name}' found.")
    eeg_df = pd.read_csv(eeg_file_name)
    print(f"First 5 rows of EEG data:\n{eeg_df.head()}")

    triggers_df = pd.read_csv(triggers_file_name)
    print(f"\nFirst 5 rows of trigger data:\n{triggers_df.head()}")
```

## Step 2: Prepare EEG Data for MNE

The raw EEG data is prepared for MNE:
* Extracting the names of all EEG channels
* Telling MNE the sampling frequency
* Arranging the EEG signal values so that each channel is correctly aligned for MNE
* Changing the unit of measurement from microvolts to standard units (volts)

```python
# all columns except 'Timestamp'
ch_names = eeg_df.columns[1:].tolist()
ch_types = ['eeg'] * len(ch_names)

# create MNE Info object
info = mne.create_info(ch_names=ch_names, sfreq=sfreq, ch_types=ch_types)

eeg_data = eeg_df.iloc[:, 1:].values.astype(np.float64).T * 1e-6

# create MNE Raw object
raw = mne.io.RawArray(eeg_data, info)

print("\nMNE Raw object created successfully")
```

## Step 3: Prepare Events (Triggers) for MNE

In this step, the trigger timestamps from the CSV file are prepared into an event array. This array acts as a precise timeline, marking exactly when each stimulus (like "happy" or "sad") appeared during the continuous EEG recording.

```python
event_id = {'happy': 1, 'sad': -1}

events = []

for index, row in triggers_df.iterrows():
    sample_index = int(row['Timestamp'] * sfreq)
    trigger_type = event_id[row['Stimulus']]
    events.append([sample_index, 0, trigger_type])

events_array = np.array(events)

print(f"Detected events:\n{events_array}")

if events_array.size > 0:
    max_sample = raw.n_times
    
    if np.max(events_array[:, 0]) >= max_sample: 
        print("Warning: Some events exceed the duration of the EEG data")
else:
    print("No events detected")
```

## Step 4: Create Epochs


* The continuous EEG data is segmented into short time windows (epochs) centered around each trigger event
* Each epoch extends from 600 ms before to 600 ms after the stimulus
* Baseline correction is applied to remove background activity and emphasize the event-related response

```python
epochs = mne.Epochs(raw, events_array, event_id=event_id, tmin=tmin, tmax=tmax,
                    preload=True, baseline=(None, 0))

print("\nMNE Epochs object created successfully")
print(f"Number of epochs: {len(epochs)}")
print(f"Epochs details:\n{epochs}")
```

## Step 5: Calculate and Plot Evoked Potentials (ERPs)

In this final step, the average brain response is calculated for each type of stimulus (happy and sad). This helps to smooth out random brain activity and highlight the specific reactions to the presented stimuli. Afterward, these averaged brain responses are visualized as wave plots.

```python
# calculate the evoked potential for each trigger type
evoked_happy = epochs['happy'].average()
evoked_sad = epochs['sad'].average()

print("\nERPs for 'happy' and 'sad' have been calculated.")

evoked_happy.plot(spatial_colors=True, gfp=True,
                  titles='ERP for HAPPY stimuli',
                  window_title='Happy ERP')

evoked_sad.plot(spatial_colors=True, gfp=True,
                titles='ERP for SAD stimuli',
                window_title='Sad ERP')

plt.show() 
```

```python
mne.viz.plot_compare_evokeds({'Happy': evoked_happy, 'Sad': evoked_sad},
                             picks='eeg', 
                             title='ERP Comparison: Happy vs Sad')

plt.show() 
```
